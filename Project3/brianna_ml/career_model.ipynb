```python
#Drop columns that are irrelevant to analysis
#Dropping date as it will interrupt one hot encoding later on
new_df = final_df.drop(columns=['Rk','G','Age','Tm','Unnamed: 5','Opp','Unnamed: 7','GS','FG%','3P%','FT%','GmSc','+/-'])
```

```python
# Import matplotlib for plotting and use magic command for Jupyter Notebooks
import matplotlib.pyplot as plt

%matplotlib inline

# Set the style
plt.style.use('fivethirtyeight')
```

```python
# Set up the plotting layout
fig, ((ax1), (ax2), (ax3), (ax4), (ax5)) = plt.subplots(nrows=5, ncols=1, figsize = (30,30))
fig.autofmt_xdate(rotation = 45)

# Field Goals 
ax1.bar(new_df['Date'], new_df['FG'])
ax1.set_xlabel('Date'); ax1.set_ylabel('Made'); ax1.set_title('Field Goals')

# Free Throws
ax2.bar(new_df['Date'], new_df['FT'])
ax2.set_xlabel('Date'); ax2.set_ylabel('Made'); ax2.set_title('Free Throw')

# Offensive Rebounds
ax3.bar(new_df['Date'], new_df['ORB'])
ax3.set_xlabel('Date'); ax3.set_ylabel('Made'); ax3.set_title('Offensive Rebounds')

# Turnovers
ax4.bar(new_df['Date'], new_df['TOV'])
ax4.set_xlabel('Date'); ax4.set_ylabel('Made'); ax4.set_title('Turnovers')

# Points
ax5.bar(new_df['Date'], new_df['PF'])
ax5.set_xlabel('Date'); ax5.set_ylabel('Made'); ax5.set_title('Personal Fouls')

plt.show
```
